```python
#Producer-Tut0-1
#Simple Producer 

topic = "Tut01"
producer = KafkaProducer(bootstrap_servers=brokers)
cnt = 1
# produce asynchronously
for _ in range(100):
    msg = f" #{cnt:5} - New Message Produced."
    producer.send(topic , msg.encode("utf-8"))
    print(msg)
    time.sleep(5)
    cnt += 1
```
